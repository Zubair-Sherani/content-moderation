# BlueSky Moderation Data Collection
**DSCI 511 — Cross-Platform Content Moderation Dataset**

This notebook collects moderation labels from BlueSky via the AT Protocol, fetches the corresponding post text, and stores everything in a local SQLite database.

### What this notebook does:
1. Sets up a SQLite database with a schema that avoids redundant records
2. Fetches all known public labelers and their label definitions
3. Queries labels from each labeler
4. Fetches the actual post text for each labeled URI
5. Stores everything — with deduplication checks at every step

### APIs used:
- `com.atproto.label.queryLabels` — fetch moderation labels
- `app.bsky.labeler.getServices` — fetch labeler metadata and definitions
- `app.bsky.feed.getPosts` — fetch post text for labeled URIs
- `app.bsky.actor.getProfiles` — fetch profile text for account-level labels

### Requirements:
```
pip install atproto requests
```

> **Note:** No BlueSky login is required. All endpoints used here are public.

---
## 1. Imports and Configuration

````python
import sqlite3
import requests
import time
import json
import logging
from datetime import datetime
from pathlib import Path

# ── Configuration ──────────────────────────────────────────────────────────
DB_PATH       = "../data/moderation.db"   # shared DB across all platform notebooks
BSKY_API_BASE = "https://public.api.bsky.app/xrpc"
REQUEST_DELAY = 0.5    # seconds between API calls — stay well under rate limits
LABEL_LIMIT   = 250    # max labels per paginated request (API max)
POST_BATCH    = 25     # getPosts accepts up to 25 URIs at once
MAX_PAGES     = 20     # safety cap — increase for a larger collection run

# Known public labelers (DID → friendly name)
# BlueSky's own moderation service + major community labelers
LABELERS = {
    "did:plc:ar7c4by46qjdydhdevvrndac" : "Bluesky Moderation",
    "did:plc:e4elbtctnfqocyfcml6h2lf7" : "Bluesky Social Team",
    "did:plc:4ugewi6aca52a62u62jccbl7" : "Community Labeler (Anti-Harassment)",
    "did:plc:cpanhictwfouvtglsayhfmnw" : "Community Labeler (Misinfo)",
}

# Ensure data directory exists
Path(DB_PATH).parent.mkdir(parents=True, exist_ok=True)

# Logging setup
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s  %(levelname)s  %(message)s",
    datefmt="%H:%M:%S"
)
log = logging.getLogger("bluesky_collector")

print("✓ Configuration loaded")
print(f"  Database : {DB_PATH}")
print(f"  Labelers : {len(LABELERS)} configured")
````

---
## 2. Database Setup

Four tables:
- `labelers` — one row per labeler, stores their metadata
- `label_definitions` — one row per label type per labeler
- `posts` — one row per unique post URI (the content)
- `labels` — one row per label event; `UNIQUE(uri, val, src)` prevents duplicates

All inserts use `INSERT OR IGNORE` so re-running the notebook never creates duplicate rows.

````python
def get_connection():
    """Return a SQLite connection with foreign keys enabled."""
    conn = sqlite3.connect(DB_PATH)
    conn.execute("PRAGMA foreign_keys = ON")
    conn.row_factory = sqlite3.Row
    return conn


def setup_database():
    """Create all tables if they don't already exist."""
    conn = get_connection()
    cur  = conn.cursor()

    # ── Labelers ───────────────────────────────────────────────────────────
    cur.execute("""
        CREATE TABLE IF NOT EXISTS labelers (
            did          TEXT PRIMARY KEY,
            name         TEXT,
            created_at   TEXT,
            fetched_at   TEXT
        )
    """)

    # ── Label definitions (what each labeler considers e.g. 'hate') ────────
    cur.execute("""
        CREATE TABLE IF NOT EXISTS label_definitions (
            id              INTEGER PRIMARY KEY AUTOINCREMENT,
            labeler_did     TEXT    NOT NULL REFERENCES labelers(did),
            identifier      TEXT    NOT NULL,   -- e.g. 'hate', 'spam'
            severity        TEXT,               -- 'alert' | 'inform' | 'none'
            blurs           TEXT,               -- 'content' | 'media' | 'none'
            default_setting TEXT,               -- 'hide' | 'warn' | 'ignore'
            description     TEXT,
            UNIQUE(labeler_did, identifier)     -- one definition per label per labeler
        )
    """)

    # ── Posts (the actual content) ─────────────────────────────────────────
    cur.execute("""
        CREATE TABLE IF NOT EXISTS bsky_posts (
            uri              TEXT PRIMARY KEY,
            cid              TEXT,
            author_did       TEXT,
            author_handle    TEXT,
            text             TEXT,
            lang             TEXT,
            post_created_at  TEXT,
            like_count       INTEGER,
            reply_count      INTEGER,
            repost_count     INTEGER,
            is_account_label INTEGER DEFAULT 0,  -- 1 if label is on account, not post
            fetched_at       TEXT
        )
    """)

    # ── Labels (the moderation events) ────────────────────────────────────
    cur.execute("""
        CREATE TABLE IF NOT EXISTS bsky_labels (
            id              INTEGER PRIMARY KEY AUTOINCREMENT,
            uri             TEXT    NOT NULL,
            cid             TEXT,
            label_val       TEXT    NOT NULL,   -- e.g. 'hate', 'spam', '!warn'
            label_src       TEXT    NOT NULL    REFERENCES labelers(did),
            is_negation     INTEGER DEFAULT 0,  -- 1 if this overrides a previous label
            labeled_at      TEXT,
            platform        TEXT    DEFAULT 'bluesky',
            UNIQUE(uri, label_val, label_src)   -- core deduplication constraint
        )
    """)

    # ── Index for fast lookups by URI ──────────────────────────────────────
    cur.execute("CREATE INDEX IF NOT EXISTS idx_bsky_labels_uri ON bsky_labels(uri)")
    cur.execute("CREATE INDEX IF NOT EXISTS idx_bsky_labels_val ON bsky_labels(label_val)")
    cur.execute("CREATE INDEX IF NOT EXISTS idx_bsky_posts_uri  ON bsky_posts(uri)")

    conn.commit()
    conn.close()
    print("✓ Database schema ready")
    print(f"  Tables: labelers, label_definitions, bsky_posts, bsky_labels")


setup_database()
````

---
## 3. Helper Functions

````python
def api_get(endpoint, params=None, labeler_did=None):
    """
    Make a GET request to the BlueSky public API.
    Optionally set the atproto-accept-labelers header to pull
    labels from a specific labeler.
    Returns the parsed JSON or None on failure.
    """
    url     = f"{BSKY_API_BASE}/{endpoint}"
    headers = {"Accept": "application/json"}

    if labeler_did:
        headers["atproto-accept-labelers"] = f"{labeler_did};redact"

    try:
        resp = requests.get(url, params=params, headers=headers, timeout=15)
        resp.raise_for_status()
        time.sleep(REQUEST_DELAY)
        return resp.json()
    except requests.HTTPError as e:
        log.warning(f"HTTP {resp.status_code} on {endpoint}: {e}")
        return None
    except requests.RequestException as e:
        log.error(f"Request failed on {endpoint}: {e}")
        return None


def now_iso():
    return datetime.utcnow().isoformat() + "Z"


def is_post_uri(uri):
    """Return True if URI points to a post record (not an account)."""
    return "app.bsky.feed.post" in uri


print("✓ Helper functions defined")
````

---
## 4. Collect Labeler Metadata and Label Definitions

Before collecting labels, we store who the labelers are and what their label definitions mean.
This is important for the comparison — the same label value `hate` can have different `severity` and `blurs` settings across labelers.

````python
def fetch_and_store_labelers():
    """
    Fetch labeler service records for all configured labelers and store
    their metadata + label definitions in the DB.
    Uses INSERT OR IGNORE — safe to re-run.
    """
    conn = get_connection()
    cur  = conn.cursor()

    dids = list(LABELERS.keys())

    # getServices accepts multiple DIDs at once
    data = api_get("app.bsky.labeler.getServices", params={
        "dids": dids,
        "detailed": "true"
    })

    if not data or "views" not in data:
        log.warning("No labeler data returned — check DIDs or network")
        conn.close()
        return

    labelers_stored     = 0
    definitions_stored  = 0

    for view in data["views"]:
        did      = view.get("creator", {}).get("did", "")
        name     = LABELERS.get(did, view.get("creator", {}).get("displayName", "Unknown"))
        created  = view.get("indexedAt", "")
        policies = view.get("policies", {})

        # ── Store labeler ─────────────────────────────────────────────────
        cur.execute("""
            INSERT OR IGNORE INTO labelers (did, name, created_at, fetched_at)
            VALUES (?, ?, ?, ?)
        """, (did, name, created, now_iso()))
        labelers_stored += cur.rowcount

        # ── Store label definitions ───────────────────────────────────────
        for defn in policies.get("labelValueDefinitions", []):
            identifier = defn.get("identifier", "")
            severity   = defn.get("severity", "")
            blurs      = defn.get("blurs", "")
            default_s  = defn.get("defaultSetting", "")

            # Pull English description if available
            description = ""
            for locale in defn.get("locales", []):
                if locale.get("lang") == "en":
                    description = locale.get("description", "")
                    break
            if not description and defn.get("locales"):
                description = defn["locales"][0].get("description", "")

            cur.execute("""
                INSERT OR IGNORE INTO label_definitions
                    (labeler_did, identifier, severity, blurs, default_setting, description)
                VALUES (?, ?, ?, ?, ?, ?)
            """, (did, identifier, severity, blurs, default_s, description))
            definitions_stored += cur.rowcount

    conn.commit()
    conn.close()

    print(f"✓ Labelers stored     : {labelers_stored} new")
    print(f"  Label definitions   : {definitions_stored} new")


fetch_and_store_labelers()
````

---
## 5. Collect Labels

For each labeler, we paginate through `com.atproto.label.queryLabels`.
The `UNIQUE(uri, label_val, label_src)` constraint in the DB means re-running this cell will skip any labels already stored.

````python
def fetch_labels_for_labeler(labeler_did, labeler_name):
    """
    Paginate through all labels from a single labeler and store them.
    Returns (new_count, skipped_count).
    """
    conn    = get_connection()
    cur     = conn.cursor()
    cursor  = None
    new     = 0
    skipped = 0
    page    = 0

    print(f"\n  Collecting from: {labeler_name}")

    while page < MAX_PAGES:
        params = {
            "uriPatterns": ["*"],   # all content
            "sources"    : [labeler_did],
            "limit"      : LABEL_LIMIT,
        }
        if cursor:
            params["cursor"] = cursor

        data = api_get(
            "com.atproto.label.queryLabels",
            params=params,
            labeler_did=labeler_did
        )

        if not data or "labels" not in data or not data["labels"]:
            break  # no more results

        for label in data["labels"]:
            uri        = label.get("uri", "")
            cid        = label.get("cid", None)
            label_val  = label.get("val", "")
            label_src  = label.get("src", labeler_did)
            is_neg     = 1 if label.get("neg", False) else 0
            labeled_at = label.get("cts", "")

            if not uri or not label_val:
                continue

            cur.execute("""
                INSERT OR IGNORE INTO bsky_labels
                    (uri, cid, label_val, label_src, is_negation, labeled_at)
                VALUES (?, ?, ?, ?, ?, ?)
            """, (uri, cid, label_val, label_src, is_neg, labeled_at))

            if cur.rowcount:
                new += 1
            else:
                skipped += 1

        conn.commit()
        page  += 1
        cursor = data.get("cursor")

        print(f"    Page {page:>2}: +{new} new labels so far", end="\r")

        if not cursor:
            break  # no more pages

    conn.close()
    print(f"    Done — {new} new, {skipped} already in DB")
    return new, skipped


def collect_all_labels():
    total_new     = 0
    total_skipped = 0

    for did, name in LABELERS.items():
        n, s = fetch_labels_for_labeler(did, name)
        total_new     += n
        total_skipped += s

    print(f"\n{'─'*45}")
    print(f"Total labels collected : {total_new} new, {total_skipped} skipped")


collect_all_labels()
````

---
## 6. Fetch Post Text for Labeled URIs

For every label in `bsky_labels` that points to a post (not an account), we fetch the full post text and store it in `bsky_posts`. We batch up to 25 URIs per API call (the endpoint's limit).

Posts already in the DB are skipped — no duplicate fetches.

````python
def fetch_and_store_posts():
    """
    Fetch post text for all labeled post URIs not yet in bsky_posts.
    Batches 25 URIs per request.
    """
    conn = get_connection()
    cur  = conn.cursor()

    # Only fetch URIs we don't already have text for
    cur.execute("""
        SELECT DISTINCT uri FROM bsky_labels
        WHERE uri LIKE '%app.bsky.feed.post%'
          AND uri NOT IN (SELECT uri FROM bsky_posts)
    """)
    uris = [row["uri"] for row in cur.fetchall()]
    conn.close()

    if not uris:
        print("✓ No new post URIs to fetch")
        return

    print(f"  Fetching text for {len(uris)} post URIs in batches of {POST_BATCH}...")

    stored  = 0
    missing = 0
    batches = [uris[i:i+POST_BATCH] for i in range(0, len(uris), POST_BATCH)]

    for batch_num, batch in enumerate(batches, 1):
        data = api_get("app.bsky.feed.getPosts", params={"uris": batch})

        if not data or "posts" not in data:
            missing += len(batch)
            continue

        conn = get_connection()
        cur  = conn.cursor()

        returned_uris = {p["uri"] for p in data["posts"]}

        for post in data["posts"]:
            uri    = post.get("uri", "")
            cid    = post.get("cid", None)
            author = post.get("author", {})
            record = post.get("record", {})

            # Extract language (first listed, if any)
            langs = record.get("langs", [])
            lang  = langs[0] if langs else None

            cur.execute("""
                INSERT OR IGNORE INTO bsky_posts
                    (uri, cid, author_did, author_handle, text, lang,
                     post_created_at, like_count, reply_count, repost_count,
                     is_account_label, fetched_at)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, 0, ?)
            """, (
                uri,
                cid,
                author.get("did"),
                author.get("handle"),
                record.get("text", ""),
                lang,
                record.get("createdAt"),
                post.get("likeCount", 0),
                post.get("replyCount", 0),
                post.get("repostCount", 0),
                now_iso()
            ))
            stored += cur.rowcount

        # Mark any URIs the API didn't return (deleted/private posts)
        for uri in batch:
            if uri not in returned_uris:
                cur.execute("""
                    INSERT OR IGNORE INTO bsky_posts
                        (uri, text, is_account_label, fetched_at)
                    VALUES (?, '[unavailable]', 0, ?)
                """, (uri, now_iso()))
                missing += cur.rowcount

        conn.commit()
        conn.close()

        print(f"    Batch {batch_num}/{len(batches)} done", end="\r")

    print(f"\n✓ Posts stored   : {stored}")
    print(f"  Unavailable    : {missing} (deleted or private)")


fetch_and_store_posts()
````

---
## 7. Fetch Profile Text for Account-Level Labels

Some labels are applied to entire accounts rather than individual posts.
For these, we fetch the profile bio and store it as the `text` field.

````python
def fetch_and_store_account_labels():
    """
    Fetch profile info for account-level labeled URIs (DIDs, not post URIs).
    These are URIs that don't contain 'app.bsky.feed.post'.
    """
    conn = get_connection()
    cur  = conn.cursor()

    # Account-level URIs look like 'at://did:plc:xxx' with no collection path
    cur.execute("""
        SELECT DISTINCT uri FROM bsky_labels
        WHERE uri NOT LIKE '%app.bsky.feed.post%'
          AND uri NOT IN (SELECT uri FROM bsky_posts)
    """)
    uris = [row["uri"] for row in cur.fetchall()]
    conn.close()

    if not uris:
        print("✓ No new account-level label URIs to fetch")
        return

    print(f"  Fetching profiles for {len(uris)} account-level labels...")

    # Extract DIDs from URIs like 'at://did:plc:xxx'
    dids = [uri.replace("at://", "").split("/")[0] for uri in uris]

    stored  = 0
    batches = [dids[i:i+25] for i in range(0, len(dids), 25)]
    uri_map = {uri.replace("at://", "").split("/")[0]: uri for uri in uris}

    for batch in batches:
        data = api_get("app.bsky.actor.getProfiles", params={"actors": batch})

        if not data or "profiles" not in data:
            continue

        conn = get_connection()
        cur  = conn.cursor()

        for profile in data["profiles"]:
            did    = profile.get("did", "")
            uri    = uri_map.get(did, f"at://{did}")

            cur.execute("""
                INSERT OR IGNORE INTO bsky_posts
                    (uri, author_did, author_handle, text, post_created_at,
                     is_account_label, fetched_at)
                VALUES (?, ?, ?, ?, ?, 1, ?)
            """, (
                uri,
                did,
                profile.get("handle"),
                profile.get("description", ""),  # bio text
                profile.get("indexedAt"),
                now_iso()
            ))
            stored += cur.rowcount

        conn.commit()
        conn.close()

    print(f"✓ Account profiles stored: {stored}")


fetch_and_store_account_labels()
````

## 8. Verify and Explore the Dataset

Quick sanity check — row counts and a preview of what we collected.

````python
import pandas as pd

conn = get_connection()

# ── Row counts ─────────────────────────────────────────────────────────────
tables = ["labelers", "label_definitions", "bsky_labels", "bsky_posts"]
print("── Table sizes ──────────────────────────")
for t in tables:
    count = conn.execute(f"SELECT COUNT(*) FROM {t}").fetchone()[0]
    print(f"  {t:<22}: {count:>6} rows")

conn.close()
````

````python
# ── Label distribution ─────────────────────────────────────────────────────
conn = get_connection()

df_labels = pd.read_sql_query("""
    SELECT
        l.label_val,
        lb.name AS labeler_name,
        COUNT(*)  AS count
    FROM bsky_labels l
    JOIN labelers lb ON l.label_src = lb.did
    GROUP BY l.label_val, lb.name
    ORDER BY count DESC
""", conn)

conn.close()
print("── Label distribution ───────────────────")
print(df_labels.to_string(index=False))
````

````python
# ── Sample of collected records (joined view) ──────────────────────────────
conn = get_connection()

df_sample = pd.read_sql_query("""
    SELECT
        p.uri,
        SUBSTR(p.text, 1, 80)   AS text_preview,
        p.lang,
        l.label_val,
        lb.name                 AS labeler,
        l.labeled_at,
        ld.severity,
        ld.blurs
    FROM bsky_labels l
    JOIN bsky_posts  p  ON l.uri        = p.uri
    JOIN labelers    lb ON l.label_src  = lb.did
    LEFT JOIN label_definitions ld
           ON ld.labeler_did = l.label_src
          AND ld.identifier  = l.label_val
    WHERE p.text IS NOT NULL
      AND p.text != '[unavailable]'
    LIMIT 10
""", conn)

conn.close()
print("── Sample records ───────────────────────")
pd.set_option("display.max_colwidth", 80)
df_sample
````

````python
# ── Posts with text available vs unavailable ───────────────────────────────
conn = get_connection()

available = conn.execute(
    "SELECT COUNT(*) FROM bsky_posts WHERE text != '[unavailable]' AND text != ''"
).fetchone()[0]
unavailable = conn.execute(
    "SELECT COUNT(*) FROM bsky_posts WHERE text = '[unavailable]' OR text = ''"
).fetchone()[0]

conn.close()
print(f"Posts with text     : {available}")
print(f"Posts unavailable   : {unavailable}  (deleted or private)")
print(f"Text coverage       : {available / max(available+unavailable, 1) * 100:.1f}%")
````

## 9. Summary

**What was collected:**
- Labeler metadata and label definitions (what each platform considers e.g. *hate*)
- All available moderation labels from configured labelers
- Post text for every labeled post URI
- Profile bio for account-level labels

**Deduplication strategy:**
- `labelers` — `PRIMARY KEY` on DID
- `label_definitions` — `UNIQUE(labeler_did, identifier)`
- `bsky_labels` — `UNIQUE(uri, label_val, label_src)`
- `bsky_posts` — `PRIMARY KEY` on URI
- All inserts use `INSERT OR IGNORE` — re-running this notebook is safe
